```python
print("=== CHEMICAL ENGINEERING BATCH LABEL ANALYZER ===")

# Input
batch = input("Enter batch label (Example: CHI-REACTOR-250L): ")

# Clean input
batch = batch.strip().upper()

print("\n--- BATCH INFORMATION ---")

# Length of string
print("Total characters:", len(batch))

# First character using indexing
print("First character:", batch[0])

# Search for separators
first_dash = batch.find("-")

if first_dash != -1:

    # Extract chemical code
    chem_code = batch[:first_dash]

    # Extract remaining part
    remaining = batch[first_dash + 1:]

    second_dash = remaining.find("-")

    if second_dash != -1:

        process = remaining[:second_dash]

        volume = remaining[second_dash + 1:]

        print("Chemical Code:", chem_code)
        print("Process Unit:", process)
        print("Batch Volume:", volume)

        # Replace operation
        clean_process = process.replace("REACTOR", "REACTOR UNIT")
        print("Updated Process:", clean_process)

        # Loop through chemical code
        print("\nCharacters in Chemical Code:")

        for letter in chem_code:
            print(letter)

        # Search operation
        if "L" in volume:
            print("\nVolume unit detected: Liters")

else:
    print("Invalid batch format!")
```
